```python
import pandas as pd
import joblib
import numpy as np
```

```python
data = pd.read_csv('WESAD_I.csv')
```

```python
data
```

```python
X = data.drop(columns=['labels'])
y = data['labels']
```

```python
def normalize_data(X):
    return (X - np.mean(X)) / np.std(X)
```

```python
X = normalize_data(X)
```

```python
X
```

```python
loaded_model = joblib.load('Oversample_model_All.pkl')
new_data_predictions = loaded_model.predict(X)
predictions_df = pd.DataFrame(new_data_predictions, columns=['predictions'])
predictions_df
```

```python
y.value_counts()
```

```python
predictions_df.value_counts()
```

```python
aligned_df = pd.concat([predictions_df, y], axis=1)
aligned_df_all = pd.concat([data,predictions_df, y], axis=1)
aligned_df
```

```python
same_values = (aligned_df['predictions'] == aligned_df['labels']).sum()
different_values = (aligned_df['predictions'] != aligned_df['labels']).sum()

print("Rovnaké predikcie:", same_values)
print("Rozdielne predikcie:", different_values)
```

```python
total_rows = len(aligned_df)
same_values_percentage = (same_values / total_rows) * 100
different_values_percentage = (different_values / total_rows) * 100

print("Rovnaké predikcie %:", same_values_percentage, "%")
print("Rozdielne predikcie %:", different_values_percentage, "%")
```
