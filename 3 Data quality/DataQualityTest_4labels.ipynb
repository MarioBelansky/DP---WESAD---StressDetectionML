Undersample

```python
data
```

```python
X = data.drop(columns=['labels'])
y = data['labels']
```

```python
X = data_without_duplicates.drop(columns=['labels'])
y = data_without_duplicates['labels']
```

```python
y.value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
under_sampler = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = under_sampler.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

```python
data_resampled = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
data_resampled
```

```python
data_resampled.describe()
```

```python
duplicate_rows = data_resampled.duplicated()
print("\nDuplikátne hodnoty:", duplicate_rows.sum())
```

```python
print("\nVizualizácia rozdelenia:")
data_resampled.hist(figsize=(10, 10))
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

def remove_outliers_zscore(df, columns, threshold=3):
    z_scores = stats.zscore(df[columns])
    abs_z_scores = abs(z_scores)
    filtered_entries = (abs_z_scores < threshold).all(axis=1)
    return df[filtered_entries]

columns_to_check = ['ACC_0', 'ACC_1', 'ACC_2', 'BVP', 'EDA', 'TEMP']
data_resampled = remove_outliers_zscore(data_resampled, columns_to_check)

data_resampled
```

```python
print("\nOutliers:")
data_resampled.boxplot(figsize=(10, 6))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\nKorelácia:")
correlation_matrix = data_resampled.corr()
plt.figure(figsize=(8, 6))
plt.matshow(correlation_matrix, cmap='coolwarm', fignum=1)
plt.colorbar()
plt.xticks(range(len(data_resampled.columns)), data_resampled.columns, rotation=90)
plt.yticks(range(len(data_resampled.columns)), data_resampled.columns)
plt.show()
```

```python
data_resampled.to_csv('WESAD_Undersample.csv', index=False)
```
