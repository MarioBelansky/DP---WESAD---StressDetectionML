```python
import pandas as pd
```

```python
data = pd.read_csv('../Data/WESAD_All_M_NN.csv')
```

```python
data
```

```python
# Odstránenie prázdnych hodnôt
data = data.dropna()
```

```python
data = data.reset_index()
```

```python
label = data['labels']
id = data['id']
```

```python
data = data.drop(columns=['labels'])
data = data.drop(columns=['id'])
data = data.drop(columns=['index'])
```

```python
data
```

```python
import pandas as pd
import numpy as np

# Define the normalization function
def normalize_data(data):
    return (data - np.mean(data)) / np.std(data)

# Define the denormalization function
def denormalize_data(normalized_data, original_data):
    return normalized_data * np.std(original_data) + np.mean(original_data)

# Normalize the data
normalized_data = normalize_data(data)

# Denormalize the data
denormalized_data = denormalize_data(normalized_data, data)
```

```python
denormalized_data
```

```python
normalized_data
```

```python
concatenated_data = pd.concat([normalized_data, id, label], axis=1)
```

```python
concatenated_data
```

```python
concatenated_data.to_csv('WESAD_All_W_N.csv', index=False)
```
