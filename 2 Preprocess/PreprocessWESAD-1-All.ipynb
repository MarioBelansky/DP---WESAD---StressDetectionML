```python
import pandas as pd

# Extract 'TEMP', 'EDA', 'BVP', 'ACC', 'labels', and 'id' from the data dictionary
TEMP = data['data']['TEMP']
EDA = data['data']['EDA']
BVP = data['data']['BVP']
ACC = data['data']['ACC']
labels = data['labels']
id = data['id']

# Find the minimum length among 'ids', 'TEMP', 'EDA', 'BVP', and 'ACC'
min_length = min(len(id), len(TEMP), len(EDA), len(BVP), len(ACC), len(labels))

# Truncate arrays to the minimum length
id = id[:min_length]
TEMP = TEMP[:min_length]
EDA = EDA[:min_length]
BVP = BVP[:min_length]
ACC = ACC[:min_length]
labels = labels[:min_length]

# Create lists to store every 4th value from TEMP and all values from EDA, BVP, and ACC
temp_values = []
eda_values = []
bvp_values = []
acc_values = []

# Iterate through each row of the arrays
for temp_row, eda_row, bvp_row, acc_row in zip(TEMP, EDA, BVP, ACC):
    # Take every 4th value from each row of TEMP and add it to the list
    temp_values.extend(temp_row[::4])
    # Take every value from each row of EDA, BVP, and ACC and add it to the respective lists
    eda_values.extend(eda_row[::4])
    bvp_values.extend(bvp_row.flatten())
    acc_values.extend(acc_row)

# Truncate the BVP array to match the length of 'temp_values', 'eda_values', and 'acc_values'
bvp_values = bvp_values[:len(temp_values)]
acc_values = acc_values[:len(temp_values)]

# Create a DataFrame using 'TEMP', 'EDA', 'BVP', 'ACC', 'labels', and 'id', removing the square brackets from 'TEMP'
df = pd.DataFrame({
    'id': id,
    'TEMP': [value[0] for value in temp_values],  # Extracting the value from the list
    'EDA': [value[0] for value in eda_values],  # Extracting the value from the list
    'BVP': bvp_values,
    'ACC': acc_values,
    'labels': labels
})
```
