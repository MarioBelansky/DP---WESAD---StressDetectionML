```python
import pandas as pd

# Extract 'TEMP', 'EDA', 'BVP', 'labels', and 'id' from the data dictionary
TEMP = data['data']['TEMP']
EDA = data['data']['EDA']
BVP = data['data']['BVP']
labels = data['labels']
id = data['id']

# Find the minimum length among 'ids', 'TEMP', 'EDA', and 'labels'
min_length = min(len(id), len(TEMP), len(EDA), len(labels))

# Truncate arrays to the minimum length
id = id[:min_length]
TEMP = TEMP[:min_length]
EDA = EDA[:min_length]
labels = labels[:min_length]

# Create lists to store every 4th value from TEMP and all values from EDA and BVP
temp_values = []
eda_values = []
bvp_values = []

# Iterate through each row of the arrays
for temp_row, eda_row, bvp_row in zip(TEMP, EDA, BVP):
    # Take every 4th value from each row of TEMP and add it to the list
    temp_values.extend(temp_row[::4])
    # Take every value from each row of EDA and add it to the list
    eda_values.extend(eda_row[::4])

# Flatten the BVP array
for bvp_row in BVP:
    bvp_values.extend(bvp_row.flatten())

# Truncate the BVP array to match the length of 'temp_values' and 'eda_values'
bvp_values = bvp_values[:len(temp_values)]

# Create a DataFrame using 'TEMP', 'EDA', 'BVP', 'labels', and 'id', removing the square brackets from 'TEMP'
df = pd.DataFrame({
    'ids': id,
    'TEMP': [value[0] for value in temp_values],  # Extracting the value from the list
    'EDA': [value[0] for value in eda_values],  # Extracting the value from the list
    'BVP': bvp_values,
    'labels': labels
})
```
